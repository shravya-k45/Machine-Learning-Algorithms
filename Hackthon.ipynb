##  splitting and scaling and the data

```python
x = df.drop(columns = 'y',axis=1)
y = df['y']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=24)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train=pd.DataFrame(x_train,columns=x.columns)
x_test=pd.DataFrame(x_test,columns=x.columns)
```

## logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
logmodel_fit=logmodel.fit(x_train,y_train)
```

```python
pred_train=logmodel.predict(x_train)
pred_test=logmodel.predict(x_test)
pred_test
```

```python
df_pred=pd.DataFrame(logmodel.predict_proba(x_train))
```

```python
df_pred['pred_train']=pred_train
```

```python
df_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,auc,roc_curve
print(classification_report(y_train, pred_train))
print(accuracy_score(y_train, pred_train))
```

```python
conf_mat_heart = confusion_matrix(y_true=y_train, y_pred=pred_train)
pd.DataFrame(conf_mat_heart)
```

```python
sns.heatmap(conf_mat_heart, annot=True, cbar=False, fmt = 'g', annot_kws={'size':25});
```

```python
# ROC and AUC curve
FPR, TPR, _ = roc_curve(y_train, pred_train)
FPR1, TPR1, _ = roc_curve(y_test,pred_test)
```

```python
auc_train = auc(FPR, TPR)
print('Train AUC: %0.3f' % auc_train)

auc_test = auc(FPR1, TPR1)
print('Test AUC: %0.3f' % auc_test)
```

```python
plt.figure(figsize=(15,7))
plt.plot(FPR, TPR, label = 'Train AUC: %0.3f'%auc_train)
plt.plot(FPR1, TPR1, label = 'Test AUC: %0.3f'%auc_test)

plt.plot([0,1],[0,1], ls = '--', color = 'black', lw = 0.75)

plt.title('ROC Curve')
plt.legend()
plt.show();
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
mse = mean_squared_error (y_test, pred_test)
rmse = np.sqrt(mse)

print('=======================================')
print ("R2 Score :" , r2_score(y_test,pred_test))
print('=======================================')
print ("RMSE :", rmse)
print('=======================================')
print ("mse :", mse)

print('=======================================')
print ("training set score :", logmodel_fit.score(x_train, y_train))
print ("test set score :", logmodel_fit.score(x_test,y_test))
print('=======================================')
```

## SVM

```python
## svm
from sklearn import svm 
from sklearn.svm import SVC
```

```python

#SVM Regression model
SVC_model = SVC(C = 1)
#SVM Regression model Fitting
SVC_model_fit = SVC_model.fit(x_train,y_train)
```

```python
y_pred_SVC = SVC_model_fit.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

mse = mean_squared_error (y_test, y_pred_SVC)
rmse = np.sqrt(mse)
print ("R2 Score :" , r2_score(y_test,y_pred_SVC))
print ("RMSE :", rmse)
print ("mse :", mse)
print ("training set score :", SVC_model_fit.score(x_train, y_train))
print ("test set score :", SVC_model_fit.score(x_test,y_test))
```

## hyper parameter tuning for svm

```python
from sklearn.model_selection import GridSearchCV
hyper_params_SVM = [ {'gamma': [1e-2, 1e-3, 1e-4],
                     'C': [1, 10, 100, 1000]}]
grid_search_SVM = GridSearchCV(estimator = SVC_model, 
                        param_grid = hyper_params_SVM,
                        cv = 5)  
grid_search_SVM_fit = grid_search_SVM.fit(x_train, y_train)    

print('best_score_ :',grid_search_SVM_fit.best_score_)
print('best_params_ :',grid_search_SVM_fit.best_params_)
print('best_estimator_ :',grid_search_SVM_fit.best_estimator_)
```

```python
cv_results_SVM = pd.DataFrame(grid_search_SVM_fit.cv_results_)
cv_results_SVM.head()
```
